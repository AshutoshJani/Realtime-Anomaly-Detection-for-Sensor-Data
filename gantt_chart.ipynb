```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

# Define tasks with (name, start_date, end_date)
tasks = [
    ("Milestone 1: Project Setup & Data Exploration and Analysis", datetime.date(2025, 6, 2), datetime.date(2025, 6, 20)),
    ("Milestone 1: Baseline Isolation Forest", datetime.date(2025, 6, 20), datetime.date(2025, 6, 30)),
    ("Milestone 1: Paper Drafting", datetime.date(2025, 6, 20), datetime.date(2025, 8, 15)),
    ("Milestone 2: RL Detector Design & Prototype", datetime.date(2025, 7, 1), datetime.date(2025, 7, 11)),
    ("Milestone 2: RL Training & Tuning", datetime.date(2025, 7, 12), datetime.date(2025, 7, 18)),
    ("Milestone 2: Evaluation & Comparison", datetime.date(2025, 7, 19), datetime.date(2025, 7, 25)),
    ("Milestone 2: Paper Polish & Internal Review", datetime.date(2025, 8, 16), datetime.date(2025, 8, 22)),
    ("Milestone 2: Final Revisions & Submission", datetime.date(2025, 8, 23), datetime.date(2025, 8, 31)),
    ("Milestone 3: Continuous-Learning MVP", datetime.date(2025, 7, 26), datetime.date(2025, 8, 1)),
    ("Milestone 3: Diagnostics & Visualization", datetime.date(2025, 8, 2), datetime.date(2025, 8, 8)),
]

# Set up the Gantt chart
fig, ax = plt.subplots(figsize=(12, 6))

for idx, (task, start, end) in enumerate(tasks):
    start_num = mdates.date2num(start)
    end_num = mdates.date2num(end)
    ax.broken_barh([(start_num, end_num - start_num)], (idx - 0.4, 0.8))

# Formatting
ax.set_yticks(range(len(tasks)))
ax.set_yticklabels([task for task, _, _ in tasks])
ax.xaxis_date()
locator = mdates.WeekdayLocator(byweekday=mdates.MO, interval=1)
formatter = mdates.DateFormatter("%b %d")
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)
plt.xticks(rotation=45)
plt.title("Accelerated Project Timeline Gantt Chart")
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

# Define tasks with (name, start_date, end_date, milestone)
tasks = [
    ("Milestone 1: Project Setup, Data Exploration & Analysis", datetime.date(2025, 6, 2), datetime.date(2025, 6, 20), 1),
    ("Milestone 1: Baseline Isolation Forest", datetime.date(2025, 6, 20), datetime.date(2025, 6, 30), 1),
    ("Milestone 1: Paper Drafting", datetime.date(2025, 6, 20), datetime.date(2025, 8, 15), 1),
    ("Milestone 2: RL Detector Design & Prototype", datetime.date(2025, 7, 1), datetime.date(2025, 7, 11), 2),
    ("Milestone 2: RL Training & Tuning", datetime.date(2025,7,12), datetime.date(2025,7,18), 2),
    ("Milestone 2: Evaluation & Comparison", datetime.date(2025,7,12), datetime.date(2025,7,25), 2),
    ("Milestone 3: Real-Time Anomaly Detection", datetime.date(2025,7,26), datetime.date(2025,8,8), 3),
    ("Milestone 3: Continuous-Learning MVP", datetime.date(2025,8,1), datetime.date(2025,8,8), 3),
    ("Milestone 2: Paper Polish, Review and Feedback", datetime.date(2025,8,5), datetime.date(2025,8,22), 2),
    ("Milestone 2: Final Revisions & Submission", datetime.date(2025,8,20), datetime.date(2025,8,31), 2),
]

# Color map per milestone
colors = {1: 'tab:blue', 2: 'tab:green', 3: 'tab:orange'}

# Create Gantt chart
fig, ax = plt.subplots(figsize=(12, 6))

for idx, (name, start, end, mstone) in enumerate(tasks):
    start_num = mdates.date2num(start)
    duration = mdates.date2num(end) - start_num
    ax.broken_barh([(start_num, duration)], (idx - 0.4, 0.8), facecolors=colors[mstone])

# Format y-axis and x-axis
ax.set_yticks(range(len(tasks)))
ax.set_yticklabels([name for name, *_ in tasks])
ax.xaxis_date()
locator = mdates.WeekdayLocator(byweekday=mdates.MO, interval=1)
formatter = mdates.DateFormatter("%b %d")
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)
plt.xticks(rotation=45)
plt.title("Accelerated Project Timeline Gantt Chart")
plt.tight_layout()
plt.legend(handles=[plt.Line2D([0], [0], color=colors[m], lw=4) for m in colors],
           labels=[f'Milestone {m}' for m in colors],
           title='Milestones', loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import WeekdayLocator, DateFormatter
import datetime

tasks = [
    ("Milestone 1: Project Setup, Data Exploration & Analysis", datetime.date(2025, 6, 2), datetime.date(2025, 6, 20), 1),
    ("Milestone 1: Baseline Isolation Forest", datetime.date(2025, 6, 20), datetime.date(2025, 6, 30), 1),
    ("Milestone 1: Paper Drafting", datetime.date(2025, 6, 20), datetime.date(2025, 8, 15), 1),
    ("Milestone 2: RL Detector Design & Prototype", datetime.date(2025, 7, 1), datetime.date(2025, 7, 11), 2),
    ("Milestone 2: RL Training & Tuning", datetime.date(2025,7,12), datetime.date(2025,7,18), 2),
    ("Milestone 2: Evaluation & Comparison", datetime.date(2025,7,12), datetime.date(2025,7,25), 2),
    ("Milestone 3: Real-Time Anomaly Detection", datetime.date(2025,7,26), datetime.date(2025,8,8), 3),
    ("Milestone 3: Continuous-Learning MVP", datetime.date(2025,8,1), datetime.date(2025,8,8), 3),
    ("Milestone 2: Paper Polish, Review and Feedback", datetime.date(2025,8,5), datetime.date(2025,8,22), 2),
    ("Milestone 2: Final Revisions & Submission", datetime.date(2025,8,20), datetime.date(2025,8,31), 2),
]

fig, ax = plt.subplots(figsize=(12, 6))

# plot the bars exactly as before
for idx, (name, start, end, mstone) in enumerate(tasks):
    start_num = mdates.date2num(start)
    duration  = mdates.date2num(end) - start_num
    ax.broken_barh([(start_num, duration)], (idx-0.4, 0.8),
                   facecolors=colors[mstone], edgecolor='black')

# --- set up weekly grid starting June 2 ---
# Major ticks on every Monday (change `byweekday=mdates.MO` to any weekday)
week_locator = WeekdayLocator(byweekday=mdates.MO, interval=1)
ax.xaxis.set_major_locator(week_locator)

# Format tick labels as “Jun 02”, etc.
ax.xaxis.set_major_formatter(DateFormatter("%b %d"))

# Turn on vertical grid lines at the major ticks
ax.grid(which='major', axis='x', linestyle='--', alpha=0.7)

# Optionally, you can also draw a vertical line at June 2 to mark the project start:
start_line = datetime.date(2025, 6, 2)
ax.axvline(mdates.date2num(start_line), color='gray', linestyle=':', linewidth=1)

# Final formatting
plt.xticks(rotation=45)
ax.set_yticks(range(len(tasks)))
ax.set_yticklabels([t[0] for t in tasks])
plt.legend(handles=[plt.Line2D([0], [0], color=colors[m], lw=4) for m in colors],
           labels=[f'Milestone {m}' for m in colors],
           title='Milestones', loc='upper left')
plt.title("Accelerated Project Timeline with Weekly Grid")
plt.tight_layout()
plt.show()
```
